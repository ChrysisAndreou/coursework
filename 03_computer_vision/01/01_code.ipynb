```python
#1_a_defineArrays

import numpy as np

M = np.array([[4, 8, 11],
              [3, 6, 2],
              [9, 7, 1],
              [10, 0, 5]])

a = np.array([3, 1, 4])
b = np.array([-2, 6, 0])
c = np.array([7, 3, 6])

# added \n\t for readable output
print ("1_a_defineArrays")
print("\tM:\n\t", np.array_str(M).replace('\n', '\n\t'))
print("\ta:", a)
print("\tb:", b)
print("\tc:", c, end="\n\n")



#1_b_dotProduct
dotProduct_ab = np.dot(a, b) # first way
# dotProduct_ab = a @ b # second way
print ("1_b_dotProduct")
print("\tdotProduct_ab:", dotProduct_ab, end="\n\n")


#1_c_a*b
elementWiseProduct_ab = a * b # first way
# elementWiseProduct_ab = np.multiply(a, b) # second way
print ("1_c_a*b")
print("\telementWise_ab:", elementWiseProduct_ab, end="\n\n")


#1_d_Ma(c^Tb)
#1_d_1_dotProduct_cb
dotProduct_cb = np.dot(c, b) #  no need to transpose the vector explicitly when using one-dimensional arrays
# dotProduct_cb = 4

#1_d_2_dotProduct_Ma
dotProduct_Ma = M@a # second way # results in (64,23,38,50)

#1_d_3_Ma(cb)
solution = dotProduct_Ma*dotProduct_cb

print("1_d_Ma(c^Tb)")
print("\tMa(cb):", solution, end="\n\n")


#1_e_magnitude a and b
# #1_e_1 : fitst way np.linalg.norm
# magnitude_a = np.linalg.norm(a) # L2 norm (Euclidean norm)
# magnitude_b = np.linalg.norm(b)
# print ("1_e_magnitude a and b")
# print("\tmagnitude_a:", magnitude_a)
# print("\tmagnitude_b:", magnitude_b, end="\n\n")

#1_e_2 : second way np.sqrt
# magnitude_a_manual = np.sqrt(np.sum(a**2))
# magnitude_b_manual = np.sqrt(np.sum(b**2))

# print("Manual magnitude_a:", magnitude_a_manual)
# print("Manual magnitude_b:", magnitude_b_manual)

#1_e_3 : third way def custom function
def manual_magnitude(vector):
    sum_of_squares = 0
    for element in vector:
        sum_of_squares += element ** 2
    return sum_of_squares ** 0.5

magnitude_a = manual_magnitude(a)
magnitude_b = manual_magnitude(b)

print("1_e_magnitude a and b")
print("\tmagnitude_a:", magnitude_a)
print("\tmagnitude_b:", magnitude_b, end="\n\n")


#1_f_normalize a and b
normalize_a = a / magnitude_a
normalize_b = b / magnitude_b
print("1_f_normalize a and b")
print("\tNormalized a:", normalize_a)
print("\tNormalized b:", normalize_b, end="\n\n")


#1_g_angle a and c
dotProduct_ac = a @ c
magnitude_c = manual_magnitude(c)

cos_theta = dotProduct_ac / (magnitude_a * magnitude_c)
angle_ac = np.arccos(cos_theta) * (180 / np.pi)  # convert to degrees

print("1_g_angle a and c")
print("\tangle_ac:", angle_ac, end="\n\n")



#1_h_perpendicular_d_to_ab
perpendicular_d_to_ab = np.cross(a, b)
print("1_h_perpendicular_d_to_ab")
print("\tperpendicular_d_to_ab:", perpendicular_d_to_ab, end="\n\n")



#1_i_orthonormal_basis_abd
# normalize_d
d = perpendicular_d_to_ab
magnitude_d = manual_magnitude(d)
normalize_d = d / magnitude_d

# calculate Q
Q = np.column_stack((normalize_a, normalize_b, normalize_d))

# calculate gramMatrix_Q
gramMatrix_Q = Q @ Q.T
# rounded_gramMatrix_Q = np.rint(gramMatrix_Q)
# print (rounded_gramMatrix_Q) # is the itendity matrix

#if AssertionError then message false
assert np.allclose(gramMatrix_Q, np.eye(3)), "orthonormal_basis_abd : false"
#if no AssertionError is raised the code will continue executing
print ("1_i_orthonormal_basis_abd")
print("\torthonormal_basis_abd : true", end="\n\n")
```

```python
# part 2 import libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np

#2_a_load/display images
# load images
image1 = cv2.imread('image1.jpg')
image2 = cv2.imread('image2.jpg')

# convert BGR to RGB for Matplotlib
image1_rgb = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB) # plt.imshow needs rgb
image2_rgb = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

# plot images
plt.figure(figsize=(10, 5)) #dimensions

# image1
plt.subplot(1, 2, 1) #1*2 grid
plt.imshow(image1_rgb)
plt.title('Image 1')
plt.axis('off') #remove x-y-axis

# image2
plt.subplot(1, 2, 2)
plt.imshow(image2_rgb)
plt.title('Image 2')
plt.axis('off')

# show
plt.show()

#2_b_display double precision and rescaled to 0-1


# double precision to allow decimals to allow rescaling (/255)
# because rescaling uint will result in zero or one (only for 255/255=1)
image1_double = image1_rgb.astype(np.float64)
image2_double = image2_rgb.astype(np.float64)

# for rescaling avoided using the built in function as instructed
# although this function will not result in the same rescaling if there is not at least one pixel with value 255
# image1_rescale = cv2.normalize(image1_double, None, 0, 1, cv2.NORM_MINMAX)


# rescale manually image to 0-1
image1_rescale = image1_double /255.0
image2_rescale = image2_double /255.0

# plot images in a 1x2 grid
plt.figure(figsize=(10, 5))

# image1
plt.subplot(1, 2, 1)
plt.imshow(image1_rescale)
plt.title('image1_rescale')
plt.axis('off')

# image2
plt.subplot(1, 2, 2)
plt.imshow(image2_rescale)
plt.title('image2_rescale')
plt.axis('off')

# show
plt.show()

# 2_c add images and re-normalize

# # the two images have the same shape
# if image1_rescale.shape == image2_rescale.shape:
#     print("same.")

# Add images of same shape
added_image = image1_rescale + image2_rescale

# Re-normalize to the range 0-1 (min-max normalization)
min_val = np.min(added_image)
max_val = np.max(added_image)
added_image_normalized = (added_image - min_val) / (max_val - min_val)

# Plot
plt.figure(figsize=(5, 5))
plt.imshow(added_image_normalized)
plt.title('Added and Re-Normalized Image')
plt.axis('off')
plt.show()

# 2_d combine two halve images
# width dimension
_, width1, _ = image1_rescale.shape # (height, width, channels)
_, width2, _ = image2_rescale.shape

# find middle of image 1 and 2
middle1 = width1 // 2
middle2 = width2 // 2

# two halfs
left_half_image2 = image2_rescale[:, :middle2]  # chooses all rows and half columns from 0-middle2
right_half_image1 = image1_rescale[:, middle1:] # chooses all rows and half columns from middle1-end

# Concatenate
combined_image = np.concatenate((left_half_image2, right_half_image1), axis=1) # axis = 1 along column, horizontal concatenation

# Plot the result
plt.figure(figsize=(5, 5))
plt.imshow(combined_image)
plt.title('Combined Image (Left from image2, Right from image1)')
plt.axis('off')
plt.show()

# 2_e_combined_images_alternate_rows
# initialize empty image with same shape as image1 or image2 (both same shape)
combined_images_alternate_rows = np.zeros_like(image1_rescale)

# Assign even rows from image1_rescale and odd rows from image2_rescale
combined_images_alternate_rows[::2, :, :] = image1_rescale[::2, :, :]  # Even rows (0, 2, 4, ...)
combined_images_alternate_rows[1::2, :, :] = image2_rescale[1::2, :, :]  # Odd rows (1, 3, 5, ...)

# Plot
plt.figure(figsize=(5, 5))
plt.imshow(combined_images_alternate_rows)
plt.title('Combined Image (Even Rows from image1, Odd Rows from image2)')
plt.axis('off')
plt.show()

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def bilinear_interpolation_resize(image, target_width, target_height):
    src_height, src_width, num_channels = image.shape

    # Calculate the ratio of the old dimensions to the new dimensions
    ratio_x = src_width / target_width
    ratio_y = src_height / target_height

    # Create coordinate grids for the new image
    grid_x = np.arange(target_width) # numpy array [0, ... target_width-1]
    grid_y = np.arange(target_height) # numpy array [0, ... target_height-1]
    # grid_x, grid_y are 2D arrays both of dimension (target_height, target_width)
    # where grid_x contains x-coordinates and grid_y contains y-coordinates
    grid_x, grid_y = np.meshgrid(grid_x, grid_y)

    # Map the coordinates back to the original image space
    #  +0.5 to go from left edge to center of the target image
    #  -0.5 because the value of a pixel center in the original image is not stored in the index_center but at index-0.5
    original_x = (grid_x + 0.5) * ratio_x - 0.5
    original_y = (grid_y + 0.5) * ratio_y - 0.5

    # x on left (floor) and right (ceil), y on down and up
    x_floor = np.floor(original_x).astype(np.int32)
    y_floor = np.floor(original_y).astype(np.int32)
    x_ceil = x_floor + 1
    y_ceil = y_floor + 1

    # Apply boundary conditions to ensure indices are within valid range of the original image to avoid IndexErrors
    x_floor = np.clip(x_floor, 0, src_width - 1) # minimum 0, and maximum src_wiedth-1
    x_ceil = np.clip(x_ceil, 0, src_width - 1)
    y_floor = np.clip(y_floor, 0, src_height - 1)
    y_ceil = np.clip(y_ceil, 0, src_height - 1)

    # pixel values of neighborns in the original image
    top_left = image[y_floor, x_floor]      # Top-left neighbor
    top_right = image[y_floor, x_ceil]      # Top-right neighbor
    bottom_left = image[y_ceil, x_floor]    # Bottom-left neighbor
    bottom_right = image[y_ceil, x_ceil]    # Bottom-right neighbor


    # Compute interpolation weights
    # x_diff and y_diff with respect to top-left neighbor
    x_diff = original_x - x_floor
    y_diff = original_y - y_floor
    weight_tl = (1 - x_diff) * (1 - y_diff)  # Top-left weight inspired from the opposite rectangle area see L04S75
    weight_tr = x_diff * (1 - y_diff)        # Top-right weight
    weight_bl = (1 - x_diff) * y_diff        # Bottom-left weight
    weight_br = x_diff * y_diff              # Bottom-right weight

    # Perform the interpolation
    # Adding None to the weights expands them to 3D (for RGB channels) to match pixel dimensions
    resized_image = (top_left * weight_tl[..., None] + top_right * weight_tr[..., None] +
                    bottom_left * weight_bl[..., None] + bottom_right * weight_br[..., None])

    # Interpolation may result in values > 255, so clip to [0, 255]
    # convert to uint8 image standard
    resized_image = np.clip(resized_image, 0, 255).astype(np.uint8)

    return resized_image

def check_resizing(img1, img2, value_thresh=1):
    """
    Check if pixel values agree for img1 and img2 under the given threshold
    :param img1: numpy.array(uint8), input resized img1
    :param img2: numpy.array(uint8), input resized img2
    :param value_thresh: int, pixel value threshold
    :return:
    None
    """
    assert img1.shape == img2.shape
    img1_reshaped = img1.reshape(-1, 3).astype(np.int32)
    img2_reshaped = img2.reshape(-1, 3).astype(np.int32)
    dist = np.abs(img1_reshaped - img2_reshaped)
    n_pixels = np.prod([s for s in img1.shape[:2]])
    dist_per_pixel = np.sum(dist, axis=1)
    n_correct_pixels = np.sum(dist_per_pixel <= img1.shape[-1] * value_thresh)
    pixel_ratio = n_correct_pixels / n_pixels
    print(f'Resizing: {pixel_ratio*100:.1f}% pixels in agreement (thr={value_thresh})')


# Load the original image
original_image = cv2.imread('image3.png')

# Retrieve the original dimensions
orig_height, orig_width = original_image.shape[:2]

# Specify the scaling factor and calculate new dimensions
scale = 5
new_width = orig_width * scale
new_height = orig_height * scale

# Perform custom bilinear interpolation resizing
custom_resized_image = bilinear_interpolation_resize(original_image, new_width, new_height)

# Use OpenCV's built-in resizing function for comparison
opencv_resized_image = cv2.resize(original_image, (new_width, new_height), interpolation=cv2.INTER_LINEAR)

# Display the images side by side
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(opencv_resized_image, cv2.COLOR_BGR2RGB))
plt.title('OpenCV Resize')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(custom_resized_image, cv2.COLOR_BGR2RGB))
plt.title('Custom Resize')
plt.axis('off')

plt.tight_layout()
plt.show()

# Compare the custom resized image with OpenCV's result
check_resizing(custom_resized_image, opencv_resized_image)


## part3

# Specify new dimensions
new_width = 160
new_height = 90

# Perform custom bilinear interpolation resizing
custom_resized_image = bilinear_interpolation_resize(original_image, new_width, new_height)

# Use OpenCV's built-in resizing function for comparison
opencv_resized_image = cv2.resize(original_image, (new_width, new_height), interpolation=cv2.INTER_LINEAR)

# Display the images side by side
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(opencv_resized_image, cv2.COLOR_BGR2RGB))
plt.title('OpenCV Resize')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(custom_resized_image, cv2.COLOR_BGR2RGB))
plt.title('Custom Resize')
plt.axis('off')

plt.tight_layout()
plt.show()

# Compare the custom resized image with OpenCV's result
check_resizing(custom_resized_image, opencv_resized_image)
```
